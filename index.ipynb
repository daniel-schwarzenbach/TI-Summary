## HГјllenoperator

::: def

Sei A eine Menge. Eine totale Funktion $О“:\mathcal{P}(A) вҶ’ \mathcal{P}(A)$ heiГҹt **HГјllenoperator** Гјber A, falls folgende Bedingungen gelten:

- Einbettung: $B вҠҶ О“(B)$ fГјr alle $B вҠҶ A$
- Monotonie: $B вҠҶ C вҮ’ О“(B) вҠҶ О“(C)$ fГјr alle $B,C вҠҶ A$
- Abgeschlossenheit: $О“(О“(B)) вҠҶ О“(B)$

:::

::: {.callout-tip collapse="true"}
#### Beispiel:

FГјr $A = в„•$ und $|$(Teilbarkeit) ist $О“(B) = \{n вҲЈ \text{es gibt } m вҲҲ B \text{ mit } n|m\}$ ein HГјllenoperator. Insbesondere gilt:

- $О“(\{ 0 \}) = в„•$
- $О“(\{24\}) = \{1,2,3,4,6,12,24\}$
- $О“(\{13\}) = \{1, 13\}$
:::

::: def


Ein HГјllenoperator $О“: \mathcal{P}(A) вҶ’ \mathcal{P}(A)$ heiГҹt **algebraisch**, falls gilt: 

FГјr alle $B вҠҶ A$ und fГјr alle $b вҲҲ О“(B)$ gibt es endliches $B_0 вҠҶ B$ mit $b вҲҲ О“(B_0)$

:::

::: {.callout-tip collapse="true"}
#### Beispiel:
Definiere $О“:\mathcal{P}(в„•) вҶ’ \mathcal{P}(в„•)$ wie folgt:

$$
О“(B) = \begin{cases}
B & \text{falls } |B| < \infty \\
в„• & \text{sonst}
\end{cases}
$$

Dann ist $О“$ ein HГјllenoperator. Es gilt $О“(в„•_+) = в„•, \; 0 вҲҲ О“(в„•_+)$

aber: $0 вҲү О“(B_0) = B_0$ fГјr endliche $B_0 вҠҶ в„•_+$

:::

### Abgeschlossenheit unter Operation $O$ 

::: def

Ein Funktion $O: A^s вҶ’ A$ heiГҹt **$s$-stellige Operation** auf $A$.

:::

::: def

Sei $O: A^s вҶ’ A$ eine $s$-stellige Operation auf $A$. Eine Menge $C вҠҶ A$ heiГҹt **abgeschlossen unter $O$** falls gilt:
$$a_1,...,a_s вҲҲ C \quad вҮ’ \quad O(a_1,...,a_s) вҲҲ C$$

:::

::: def

Seien $O_i: A^{s_i} вҶ’ A$ fГјr $i вҲҲ \{1,...,k\}$ Operationen auf $A$, $B вҠҶ A$. Dann ist $О“_{O_1,...,O_k}(B)$ die kleinste Menge, die $B вҠҶ A$ enthГӨlt und abgeschlossen ist unter $O_1,...,O_k$ d.h.

$$О“_{O_1,..,O_k} (B) = вӢӮ \{C вҲЈ B вҠҶ C, \; C \text{ abgeschlossen unter } O_1, ..., O_k\}$$
$О“_{O_1,..,O_k}$ heisst der durch $O_1,...,O_k$ erzeugte HГјllenoperator.

:::

::: {.callout-tip collapse="true"}
#### Beispiel:

- $О“_{\text{sum}}(\{1\}) = в„•_+$ da $1+1=2, 1+2=3, 2+2=4,...$
- $О“_{\text{prod}}(\{1\}) = \{1\}$ da $1вӢ…1=1$
- $О“_{\text{sum,prod}}(\{2, 3\}) = \{xвҲҲв„• вҲЈ x = 2^a вӢ… 3^b \text{ mit } a+b вүҘ 1\}$

:::


::: prop

Anmerkung:

$О“_{O_1,..,O_k}$ ist ein **algebraischer** HГјllenoperator.
:::

### Operatoren auf $\textbf{FUNK}(A)$

:::def

Sei A eine beliebige Menge. Definiere dann die Menge der **Operationen auf A** als:
$$\textbf{FUNK}(A) := \{ПҶ вҲЈ вҲғ nвҲҲв„• \text{ mit } ПҶ: A^n вҶ’ A\}$$

:::

$\textbf{FUNK}(в„•)$ ist fГјr die Theoretische Informatik von fundamentaler Bedeutung. Unter anderem ist $\textbf{TM} вҠҶ \textbf{FUNK}(в„•)$.


:::def

Es seien $ПҶ: A^n вҶ’ A, \; ПҲ:A^m вҶ’ A \quad \in \textbf{FUNK}(A)$

- zyklische Vertauschung:
    - $\text{ZV}(ПҶ): A^n вҶ’ A$
    - $\text{ZV}(ПҶ)(x_1,...,x_n) = ПҶ(x_2,...,x_n,x_1)$
- vertauschung der letzten Variable:
    - $\text{LV}(ПҶ): A^n вҶ’ A$
    - $\text{LV}(ПҶ)(x_1,...,x_n) = ПҶ(x_1,...,x_{n-2},x_n,x_{n-1})$
- Identifikation der letzten Variablen:
    - $\text{ID}(ПҶ): A^{n-1} вҶ’ A$
    - $\text{ID}(ПҶ)(x_1,...,x_{n-1}) = ПҶ(x_1,...,x_{n-1},x_{n-1})$
- Substitution an letzter Stelle:
    - $\text{SUB}(ПҶ, ПҲ): A^{n-1+m} вҶ’ A$
    - $\text{SUB}(ПҶ, ПҲ)(x_1,...,x_{n-1},y_1,...,y_m) = ПҶ(x_1,...,x_{n-1}, ПҲ(y_1,...,y_m))$

:::

::: {.callout-tip collapse="true"}

#### Beispiel:
$$
\begin{aligned}
\text{ID}(\text{ZV}(\text{ZV}(\text{SUB}(\text{sum},\text{md})))) (x,y) & = \text{ZV}(\text{ZV}(\text{SUB}(\text{sum},\text{md}))) (x,y,y) \\
& = \text{ZV}(\text{SUB}(\text{sum},\text{md})) (y,y,x) \\
& = \text{SUB}(\text{sum},\text{md}) (y,x,y) \\
& = \text{sum}(y,\text{md}(x,y)) \\
& = y + (x вҲё y) \\
& = \max\{x,y\}
\end{aligned}
$$

mithin gilt dann:

$$
\max вҲҲ О“_{\text{ZV}, \text{ID}, \text{SUB}}(\{\text{sum}, \text{md}\})
$$
:::

## Random-Access-Maschinen

:::def

- Die Steuereinheit enthГӨlt das Programm und fГјhrt es aus. Ein Programm besteht aus einer endlichen Liste von nummerierten Befehlen.
- Das Befehlsregister enthГӨlt die Nummer des aktuellen Befehls im Programm.
- Jedes der unendlich vielen (Daten-)Register `R0, R1, R2, ...` enthГӨlt eine natГјrliche Zahl (gespeichert in dyadischer Darstellung). Die Nummer `i` von Register `Ri` heiГҹt Adresse.

Bezeichnungen:
```julia
<Ri>      # inhalt von Ri, d.h. R[i]
<BR>     # inhalt von BR вүғ *BR
<R> := x # Register R bekommt neuen Inhalt x, dh 
```

Befehle, Wirkung
```julia
# Transportbefehle
Ri вҶҗ Rj     # <Ri> := <Rj>
Ri вҶҗ RRj    # <Ri> := <R<Rj>>
RRi вҶҗ Rj    # R[R[i]] := R[j]
# Arithmetische Befehle
Ri вҶҗ k         # R[i] = k
Ri вҶҗ Rj + Rk   # R[i] =  R[j] + R[k]
Ri вҶҗ Rj - Rk   # R[i] =  R[j] - R[k]
# Sprungbefehle
IF Rj = 0 THEN GOTO x  # falls <Rj> = 0, dann gehe zu Befehl mit Nummer x
IF Rj > 0 THEN GOTO x  # falls <Rj> > 0, dann gehe zu Befehl mit Nummer x
GOTO x                 # gehe zu Befehl mit Nummer x
```

## Turingmaschinen



:::def
### Definition: **Turingmaschinen**

Eine Turingmaschine mit $k$-BГӨndern ist ein Tupel $M = (ОЈ, Z, f, zвӮҖ, z_1)$ bestehend aus:

- einem Bandalphabet $ОЈ$,
- eine Menge von ZustГӨnden $Z$,
- einer ГңberfГјhrungsfunktion $f : Z Г— ОЈ^k \; вҶ’ \; Z Г— ОЈ^k Г— \{L, 0, R\}^k$,
- einem Startzustand $zвӮҖ вҲҲ Z$,
- einem Endzustand $z_1 вҲҲ Z$.

Eine Turingmaschine besteht aus folgenden Komponenten:

- $k$ (mit $k вүҘ 1$) beidseitig unendliche, in Zellen unterteilte BГӨnder; in jeder Zelle steht genau ein Symbol aus dem Bandalphabet $ОЈ$, wobei das Leerzeichen $в–Ў вҲҲ ОЈ$ zeigt an, dass in der Zelle eigentlich nichts steht
- fГјr jedes Band ein Schreib- und Lesekopf (oder kurz einfach Kopf), der sich von Zelle zu Zelle bewegen kann, um Inhalte der Zellen zu lesen und zu verГӨndern
- eine Steuereinheit, die sich in einem Zustand $z$ aus einer endlichen Zustandsmenge $Z$ befindet, die Гјber die KГ¶pfe Bandinhalte liest und die AktivitГӨten entsprechend steuert

das Verhalten pro Takt wird durch eine (totale) ГңberfГјhrungsfunktion:
$$
f : Z Г— ОЈ^k \quad вҶ’ \quad Z Г— ОЈ^k Г— \{L, 0, R\}^k
$$
beschrieben; die Beschreibung ist unabhГӨngig von Taktnummer und Kopfpositionen.

:::


::: def

### *Definition:* **Turing-Berechenbarkeit**



1. Seien $M = (ОЈ, Z, f, z_0, z_1)$ eine Turingmaschine, $ОЈ_1 вҠҶ ОЈ вҲ– \{в–Ў,*\}, ОЈ_2 вҠҶ ОЈ вҲ– \{в–Ў\}, n вҲҲ N$. Eine Funktion $ПҲ : (ОЈ^вҲ—_1)^n вҶ’ ОЈ^вҲ—_2$ heiГҹt von der Turingmaschine $M$ **berechnet**, falls fГјr alle $x_1, ... , x_n вҲҲ ОЈ^*_1$ gilt:

(a) Ist $ПҲ(x_1, ... , x_n)$ definiert, so gilt: Startet $M$ in $S(z_0, x_1 вҲ— x_2 вҲ— В·В·В· вҲ— x_n)$, so stoppt $M$ in $S(z_1, ПҲ(x_1, ..., x_n))$.

(b) Ist $ПҲ(x_1, ... , x_n)$ nicht definiert, so gilt: Startet $M$ in $S(z_0, x_1вҲ—x_2вҲ—В·В·В·вҲ—x_n)$, so stoppt $M$ nicht oder $M$ stoppt aber nicht in einer Situation der Form $S(z_1, y)$ mit $y вҲҲ (ОЈ вҲ– \{в–Ў\})^*$.

<center>

![](./img/Endzustand.jpg){width=50%}

</center>

2. Eine Funktion $ПҲ : (ОЈ^*_1)^n вҶ’ ОЈ^*_2$ heiГҹt **Turing-berechenbar**, falls es eine Turingmaschine gibt, die $ПҲ$ berechnet.

3. Eine Funktion $ПҶ : в„•^n вҶ’ в„•$ heiГҹt **Turing-berechenbar**, falls sie in dyadischer Kodierung Turing-berechenbar ist, d.h. falls die durch $ПҲ(x_1, ... , x_n) = \text{dya}(ПҶ(\text{dya}^{-1}(x_1), ..., \text{dya}^{-1}(x_n)))$ definierte Funktion $ПҲ : (\{1, 2\}^вҲ—)^n вҶ’ \{1, 2\}^*$ Turing-berechenbar ist.

<center>

</center>

$$
\textbf{TM} = \{ ПҶ \mid ПҶ : в„•^n вҶ’ в„• \text{ ist Turing-berechenbar}  \}
$$

### Definition: **Primitive Rekursion**

1. Es seien $ПҶ : в„•^n вҶ’ в„•$ und $ПҲ : в„•^{n+2} вҶ’ в„•$ Funktionen. Dann ist die Funktion $\text{PR}(ПҶ, ПҲ) : в„•^{n+1} вҶ’ в„•$ definiert durch:

$$
\text{PR}(ПҶ, ПҲ)(x_1,...,x_n,y) = \begin{cases}
ПҶ(x_1,...,x_n) & \text{falls }y = 0 \\
ПҲ(x_1,...,x_n,y-1,\text{PR}(ПҶ, ПҲ)(x_1,...,x_n,y-1)) & \text{sonst}
\end{cases}
$$

2. $\textbf{PRIM} = О“_{\text{ZV},\text{ID},\text{SUB}, \text{LV}, \text{PR}}(\{\text{S}, \text{C}^0_0, \text{I}^2_1\})$.

3. Eine Funktion $f : в„•^n вҶ’ в„•$ heiГҹt **primitiv-rekursiv**, falls $f вҲҲ \textbf{PRIM}$.

:::

:::prop

Die Funktion $I^n_m$, $C^n_m$, sum, md, prod, div, exp, prim und prex sind alle **primitiv-rekursiv**.

:::

:::prop

### Methode zum Erstellen/Erkennen von primitiv-rekursiven Funktionen

Wir suchen eine Funktion $f : в„•^n вҶ’ в„•$ die primitiv-rekursiv ist, d.h. $f вҲҲ \textbf{PRIM}$. DafГјr starten wir mit dem Induktionsanfang ($y=0$) und finden ein passendes $ПҶ : в„•^n вҶ’ в„•$ so dass gilt:
$$
f(x_1,...,x_n,0) = ПҶ(x_1,...,x_n)
$$

und fГјr den Induktionsschritt ($y>0$) finden wir eine Funktion $ПҲ : в„•^{n+2} вҶ’ в„•$ so dass gilt:
$$
\begin{aligned}
f(x_1,...,x_n,y) & = ПҲ(x_1,...,x_n,y-1,f(x_1,...,x_n,y-1)) \\
& = \text{PR}(ПҶ, ПҲ)(x_1,...,x_n,y) вҲҲ \textbf{PRIM}
\end{aligned}
$$
wobei die Induktionsvoraussetzung $f(x_1,...,x_n,y-1) = \text{PR}(ПҶ, ПҲ)(x_1,...,x_n,y-1)$ ist.

N, M = 20_000, 20_000
Ak::Matrix{Int} = fill(0, N, M) # 0 indicates uncomputed 

# wrapper function of the Ak Matrix to handle base cases
function Akw(m::Int, n::Int)::Int
    # base cases
    if m == 0
        return n + 1
    elseif n == 0
        return m
    else
        return Ak[m, n]
    end
end

# tries to fill the Ak matrix at position (m, n) if the dependent values are already computed
function try_compute_Ak(m::Int, n::Int)::Nothing
    # if the value is already computed, do nothing
    if Ak[m, n] != 0
        return
    # if the dependent value is not computed, do nothing
    elseif Akw(m,n-1) != 0 && Akw(m-1, Akw(m, n-1)) != 0
        Ak[m, n] = Akw(m-1, Akw(m, n-1))
        println("Computed A($m, $n) = ", Ak[m, n])
    end
    return

    while x <= n
        i = i + 1
        x = x + i
    end
    x -= i
    i -= 1
    y = n - x
    return (i-y, y)

    while (Akw(m, n) == 0)

            x, y = nxn(i)
            try_compute_Ak(x+1, y+1)
        end
        e += 1
    end
    return Akw(m, n)

$Ојx(E(x)) = \arg \min_{x вҲҲ в„•} E(x)$, wobei $E(x)$ eine Aussage Гјber $x$ ist. Also $Ојx(E(x))$ ist die kleinste natГјrliche Zahl $x$, fГјr die $E(x)$ wahr ist. Wenn es kein $x$ gibt, fГјr das $E(x)$ wahr ist, dann ist $Ојx(E(x))$ undefiniert.

### **Das Halteproblem**

Wir kГ¶nnen jedes RAM programm $M_x$ als eine natГјrliche Zahl $x$ kodieren.



| **Symbol**    | R   | G   | I   | вҶ’    | +   | -   | =   | вү     |
|---------------|-----|-----|-----|------|-----|-----|-----|------|
| **Kodierung** | 111 | 112 | 121 | 122  | 211 | 212 | 221 | 222  |

![](./img/GГ¶delisierung.png)

ist eine natГјrliche Zahl keine gГјltige RAM, so wird sie als `STOP` kodiert.

somit kГ¶nnen wir jeder NatГјrlichen Zahl $x$ eine RAM $M_x$ zuordnen.

Diesen Vorgang nennt man auch **GГ¶delisierung**.

Das **Halteproblem** ist nicht nur fГјr uns Informatiker ein fundamentales Problem, sondern auch fГјr die Mathematik selbst, da es die Grenzen der Beweisbarkeit aufzeigt und zeigt das Mathematik kein abgeschlossenes System ist.

{{< video https://youtu.be/HeQX2HjkcNo?si=H4FQupNvg2lf6zRh&t=814 >}}

- Wichtige Stellen: **13:34 - 20:52** und **21:19 - 27:17**

:::def

Das Hilbertsche Programm, initiiert von David Hilbert zu Beginn des 20. Jahrhunderts, war ein ehrgeiziger Versuch, die gesamte Mathematik auf ein sicheres, formales Fundament zu stellen. Es zielte darauf ab, die Grundlagenkrise der Mathematik zu lГ¶sen, indem die Mathematik in ein formales System ГјberfГјhrt wird, fГјr das drei **Haupteigenschaften** gefordert wurden:

1. **Konsistenz (Widerspruchsfreiheit):** In der Mathematik gibt es keine Aussage $A$, so dass sowohl $A$ als auch $\neg A$ beweisbar sind.

2. **VollstГӨndigkeit:** FГјr jede Aussage $A$ in der Mathematik gilt: entweder $A$ oder $\neg A$ ist beweisbar.

3. **Entscheidbarkeit:** Gibt es ein Verfahren, um zu entscheiden, ob eine Aussage $A$ aus den Axiomen eines formalen Systems beweisbar ist?

:::

:::prop

Proposition zu Hilberts Haupteigenschaften:

1. Es gibt kein formales System, das zeigen kann, dass die Mathematik konsistent ist, wenn sie tatsГӨchlich konsistent ist.

2. Kein formales System, das die Mathematik beschreibt, ist vollstГӨndig.

3. Es gibt kein Verfahren, um zu entscheiden, ob eine Aussage $A$ aus den Axiomen eines formalen Systems beweisbar ist.

:::

:::prop

### Der Zusammenhang zwischen Entscheidbarkeit und Beweisbarkeit

#### 1. GГ¶delisierung als BrГјcke

Durch **GГ¶delisierung** kann jede mathematische Aussage als natГјrliche Zahl kodiert werden. Damit entspricht:

$$A \subseteq \mathbb{N} \quad \Longleftrightarrow \quad \text{eine Menge von mathematischen Aussagen}$$

Die Frage вҖһist $x \in A$?" ist ГӨquivalent zur mathematischen Frage вҖһgilt Eigenschaft $A$ fГјr $x$?".

#### 2. Was bedeutet Entscheidbarkeit fГјr Beweisbarkeit?

$$A \text{ ist entscheidbar} \quad \Longleftrightarrow \quad \forall x \in \mathbb{N}: \text{ entweder вҖһ}x \in A\text{" oder вҖһ}x \notin A\text{" ist beweisbar}$$

- Ein Algorithmus, der $A$ entscheidet, liefert implizit fГјr **jede** Instanz einen Beweis - entweder einen positiven oder einen negativen. 
- Unentscheidbarkeit heiГҹt: es gibt Aussagen, die **weder beweisbar noch widerlegbar** sind.

## Laufzeit von Algorithmen

Sei $M$ ein Algorithmus der die totale Funktion $f: (ОЈ^*)^m вҶ’ ОЈ^*$ berechnet.

$$t_M: (ОЈ^*)^m вҶ’ в„•$$

### Definition: **Laufzeit**

Sei $t: в„• вҶ’ в„•$ eine totale Funktion.

1. Algorithmus M **berechnet** $f:(ОЈ^*)^m вҶ’ ОЈ^*$ in der Zeit $t$ falls gilt:
    - $M$ berechnet $f$
    - $t_M(x_1, ..., x_m) вүӨ t(|x_1|, ..., |x_m|)$

2. Ein Algorithmus $M$ berechnet $f : (ОЈ^вҲ—)^m вҶ’ ОЈ^вҲ—$ in der Zeit $\mathcal{O}(t)$, falls es ein $c > 0$ gibt, sodass $M$ die Funktion $f$ in der Zeit $c В· t(n) + c$ berechnet.

3. Ein Algorithmus $M$ entscheidet $A вҠҶ (ОЈ^вҲ—)^m$ in der Zeit $t$ (bzw. $\mathcal{O}(t)$), falls $M$ die charakteristische Funktion $c_A$ in der Zeit $t$ (bzw. $\mathcal{O}(t)$) berechnet.

## $\textbf{NP}$-vollstГӨndige Mengen

{{< video https://www.youtube.com/watch?v=W9G_1xG77LE&list=PLlwsleWT767dnN25K_QgvdKkovK_t4K6-&index=6  >}}

:::def

Sein $AвҠҶОЈ^*_1$ und $BвҠҶОЈ^*_2$ Mengen. Falls es eine totale Funktion $f:ОЈ^*_1 вҶ’ ОЈ^*_2$ mit $f вҲҲ \textbf{FP}$ gibt
$$
x вҲҲ A \quad вҹ· \quad f(x) вҲҲ B
$$
bzw.
$$
C_A(x) = C_B(f(x))
$$
gibt, so sagen wir: $A$ ist **in Polynomialzeit reduzierbar** auf $B$ bzw. 
$$
A вүӨ^p_m B
$$

:::

:::prop

1. Die Relation $вүӨ_m^p$ ist reflexiv und transitiv.
2. FГјr alle Mengen $A$ und $B$ gilt: $A вүӨ_m^p B \quad \Leftrightarrow \quad \overline{A} вүӨ_m^p \overline{B}$

:::


:::prop

$\textbf{P}$ und $\textbf{NP}$ sind abgeschlossen unter  $вүӨ_m^p$. Daher

1. Ist $A вүӨ_m^p B$ und $B вҲҲ \textbf{P}$, so ist $A вҲҲ \textbf{P}$.
2. Ist $A вүӨ_m^p B$ und $B вҲҲ \textbf{NP}$, so ist $A вҲҲ \textbf{NP}$.

:::

:::prop

Sei $B вҠҠ в„•$ und $B вү  вҲ…$ (nicht-trivial). Dann gilt: $A вүӨ_m^p B$ fГјr alle $A вҲҲ \textbf{P}$.

:::

:::: def
### Definition: **NP-vollstГӨndig**

Eine Menge $B$ heiГҹt NP-vollstГӨndig, falls $B вҲҲ NP$ und $A вүӨ^p_m B$ fГјr alle $A вҲҲ \textbf{NP}$.
::::

:::prop

Seien $A, B вҠҶ в„•$.

1. Sind $A$ und $B$ $\textbf{NP}$-vollstГӨndig, so gilt $A вүЎ^p_m B$.
2. Ist $A$ $\textbf{NP}$-vollstГӨndig und $A вҲҲ \textbf{P}$, so gilt $\textbf{P} = \textbf{NP}$.
3. Ist $B вҲҲ \textbf{NP}$, ist $A$ $\textbf{NP}$-vollstГӨndig und $A вүӨ^p_m B$, so ist $B$ $\textbf{NP}$-vollstГӨndig.

:::

:::prop

**Anmerkung**: Ein Algorithmus, der eine NP-vollstГӨndige Menge entscheidet, kann von einer Maschine mit der sich Programme beliebig parallelisieren lassen, in polynomieller Zeit entschieden werden.

Probleme in EXP-vollstГӨndig kГ¶nnen hingegen nicht parallelisiert werden, und benГ¶tigen auch auf einer Maschine mit unendlich vielen Prozessoren exponentielle Zeit.

:::

:::prop

Folgende Mengen sind **NP**-vollstГӨndig:

- $\text{HK} = \{G \mid G \text{ ist ein hamiltonischer Graph }\}$

Ein Graph $G$ heiГҹt hamiltonisch, wenn es einen Pfad gibt, der durch alle Knoten genau einmal geht und am Ende wieder zum Startknoten zurГјckkehrt.

- $\text{SOS} = \{ (a_1, ..., a_m, b) \mid \text{es gibt eine Teilmenge } I \subseteq \{1, ..., m\} \text{ mit } \sum_{i \in I} a_i = b \}$

Die SOS-Menge besteht aus Tupeln von Zahlen, fГјr die es eine Teilmenge gibt, deren Summe genau $b$ ergibt.

- $3\text{-FARB} = \{G \mid G \text{ ist ein Graph, der mit } 3 \text{ Farben fГӨrbbar ist }\}$

Eine KnotenfГӨrbung eines Graphen $G$ mit 3 Farben ist eine Zuordnung von Farben zu den Knoten von $G$, sodass benachbarte Knoten unterschiedliche Farben haben. Ein Graph ist 3-fГӨrbbar, wenn es eine solche FГӨrbung mit hГ¶chstens 3 Farben gibt.

- $3\text{-SAT} = \{\phi \mid \phi(x_1, ..., x_n) \text{ ist eine erfГјllbare Formel in konjunktiver Normalform mit hГ¶chstens 3 Literalen pro Klausel }\}$

Eine Formel in konjunktiver Normalform (CNF) ist eine Konjunktion von Klauseln, wobei jede Klausel eine Disjunktion von Literalen ist. Ein Literal ist entweder eine Variable oder die Negation einer Variable. Eine Formel ist erfГјllbar, wenn es eine Belegung der Variablen gibt, die die Formel wahr macht.

z.B die Formel $(x_1 вҲЁ В¬x_2 вҲЁ x_3) вҲ§ (В¬x_1 вҲЁ x_2 вҲЁ В¬x_3)$ ist in 3-CNF und hat 2 Klauseln mit jeweils 3 Literalen.

- $\text{PUZZLE} = \{ (m, k, t, R, T_1, \ldots, T_k) \mid m, k, t \ge 1,\ R \in (\Sigma^m)^{4t},\ T_1, \ldots, T_k \in (\Sigma^m)^4 \text{ und } R \text{ kann mit Puzzleteilen der Typen } T_1, \ldots, T_k \text{ gefГјllt werden} \}$

:::


:::def

### Definition: **P-vollstГӨndig**

Eine Menge $B$ heiГҹt P-vollstГӨndig, falls $B вҲҲ P$ und $A вүӨ^p_m B$ fГјr alle $A вҲҲ \textbf{P}$.

:::

:::prop

**Anmerkung**: Ein Algorithmus, der eine P-vollstГӨndige Menge entscheidet, ist nicht parallelisierbar.
:::

:::prop

Folgende Mengen sind **P**-vollstГӨndig:

- $\text{PATH} = \{(G, s, t) \mid G \text{ ist ein gerichteter Graph mit einem Pfad von } s \text{ nach } t\}$

Ein gerichteter Graph $G$ besteht aus einer Menge von Knoten und gerichteten Kanten zwischen diesen Knoten. Ein Pfad von $s$ nach $t$ ist eine Sequenz von Knoten, beginnend mit $s$ und endend mit $t$, wobei jede aufeinanderfolgende Paar von Knoten durch eine gerichtete Kante verbunden ist.

- $\text{KAP} = \{(K, x) \mid K \text{ ist ein boolescher Schaltkreis und } K(x) = 1\}$

Das **Kreisauswertungsproblem (KAP)** (engl. *Circuit Value Problem*, CVP) fragt, ob ein gegebener boolescher Schaltkreis $K$ bei einer gegebenen Eingabe $x$ den Wert $1$ ausgibt. Ein boolescher Schaltkreis besteht aus AND-, OR- und NOT-Gattern sowie Eingabe- und Ausgabeknoten. KAP gilt als das kanonische P-vollstГӨndige Problem.

- $\text{HORNSAT} = \{\phi \mid \phi \text{ ist eine erfГјllbare Horn-Formel}\}$

Eine **Horn-Formel** ist eine Konjunktion von Horn-Klauseln, wobei jede Horn-Klausel eine Disjunktion von Literalen mit **hГ¶chstens einem positiven Literal** ist. HORNSAT fragt, ob eine gegebene Horn-Formel erfГјllbar ist. z.B. die Formel $(В¬x_1 вҲЁ В¬x_2 вҲЁ x_3) вҲ§ (В¬x_1 вҲЁ x_4) вҲ§ (В¬x_2 вҲЁ x_5) вҲ§ (В¬x_3 вҲЁ В¬x_4 вҲЁ В¬x_5)$ ist eine Horn-Formel, da jede Klausel hГ¶chstens ein positives Literal enthГӨlt.

:::
